# Basic Cleaning

```python
df.columns = df.columns.str.lower().str.replace(" ","_")
df.head()
```

```python
df["attrition_flag"] = df["attrition"].map({"Yes":1, "No":0})
```

```python
df["overtime_flag"] = df["overtime"].map({"Yes":1, "No":0})
```

# Feature Engineering
Creating new meaningful columns.

```python
def tenure_band(years):
    if years <= 2:
        return "0-2 Years"
    elif years <= 5:
        return "3-5 Years"
    elif years <= 10:
        return "6-10 Years"
    else:
        return "10+ Years"

df["tenure_band"] = df["yearsatcompany"].apply(tenure_band)
```

```python
df["income_group"] = pd.qcut(df["monthlyincome"], q=3, labels=["Low", "Medium", "High"])
```

```python
df["age_group"] = pd.cut(df["age"], bins = [18,30,40,50,60], labels = ["18-30", "31-40", "41-50", "51-60"])
```

```python
attrition_rate = df["attrition_flag"].mean() * 100
print(f"Overall Attrition Rate: {attrition_rate: .2f}%")
```

# EDA

```python
dept_attrition = df.groupby("department")["attrition_flag"].mean().sort_values(ascending=False)*100
dept_attrition.plot(kind="bar")
plt.title("Attrition Rate by Department")
plt.ylabel("Attrition Rate (%)")
```

```python
job_attrition = df.groupby("jobrole")["attrition_flag"].mean().sort_values(ascending=False)*100
job_attrition.plot(kind="bar")
plt.title("Attrition Rate by Job Role")
plt.ylabel("Attrition Rate(%)")
plt.xticks(rotation=45)
```

```python
sns.barplot(x="overtime", y="attrition_flag", data=df)
plt.title("Overtime v/s Attrition")
plt.ylabel("Attrition Rate")
```

```python
df.groupby("overtime")["attrition_flag"].mean()*100
```

```python
sns.boxplot(x="attrition", y="monthlyincome", data=df)
plt.title("Income Distribution by Attrition")
```

```python
sns.barplot(x="tenure_band", y="attrition_flag", data=df)
plt.title("Attrition by Tenure Band")
plt.ylabel("Attrition Rate")
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Correlation Heatmap")
```

# Key Findings from EDA
1. Department with Highest Attrition → Sales

Meaning:
Sales department is experiencing the highest employee turnover.

Possible reasons:

- High pressure targets
- Performance-based stress
- Incentive variability

2. Most Risky Job Role → Sales Representative

This indicates:
Front-line sales employees are more likely to leave.

3. Overtime Increases Attrition → Yes

If overtime employees have significantly higher attrition:
It suggests work-life imbalance is affecting retention.

4. Low Income Employees Leave More → Yes

This shows:
- Salary competitiveness may be an issue.
- Compensation review could reduce attrition.

5. Highest Attrition in Tenure Band → 0–2 Years

It means:
Most employees leave early in their career at the company.

This suggests:
- Weak onboarding
- Poor early engagement
- Misaligned expectations
